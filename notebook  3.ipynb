```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df2 = pd.read_csv("Data_HW7.csv")
df2.head()
```

```python
df2['Delta_tc_ft'] = df2['Delta_tc_m']*0.3048
df2.head()
```

### 1) Compressional velocity (vc) in m/sec

$$V_c = compressional velocity (ft/sec) =  1/ \Delta t * 1000000$$ 

```python
def Vc(tc_m):
    Vc = (1/tc_m)*1000000
    return Vc
```

```python
df2['Vc_m/sec'] = Vc(df2['Delta_tc_m'])
df2.head()
```

### 2) Shear transit time (Dtshear) in msec/m

Dt shear = ( Dt comp - 20.408922 ) / 0.422373

```python
def Dtshear(tc_ft):
    Dtshear = (tc_ft-20.408922)/(0.422373)
    return Dtshear
```

```python
df2['Dtshear_ft'] = Dtshear(df2['Delta_tc_ft'])
df2.head()
```

```python
df2['Dtshear_m'] = (df2['Dtshear_ft'])/0.3048
df2.head()
```

### 3) Shear velocity (vs) in msec/m

$$\Delta t_s = shear transit time (\mu sec/ft) = 1000000 / V_s$$

```python
def Vs(Dtshear):
    Vs = (1/Dtshear)*1000000
    return Vs
```

```python
df2['Vs_m/sec'] = Vs(df2['Dtshear_m'])
df2.head()
```

### 4) Dynamic Poisson's ratio (PR)

$$ PR = (0.5V_c^2 - V_s^2)/(V_c^2 - V_s^2) $$

```python
def PR(Vc,Vs):
    PR = ((0.5*Vc*Vc) - (Vs**2))/((Vc**2)-(Vs**2))
    return PR
```

```python
df2['PR']= PR(df2['Vc_m/sec'],df2['Vs_m/sec'])
df2.head()
```

#### Static Poisson's Ratio

Static Poisson's ratio = Dynamic PR x 0.9857

```python
df2['PR_static'] = 0.9857 * df2['PR']
df2.head()
```

### 5) Dynamic Shear Modulus ( G or SM )

$$ G = 1.34 * 10^{10}\rho_b / \Delta t_s^2 $$
$$ \rho_b  = bulk density  $$ 

```python
def G(RHOB, Dtshear):
    G = ((1.34*10000000000*RHOB)/1000)/(Dtshear**2)
    return G
```

```python
df2['G_psi']= G(df2['RHOB'],df2['Dtshear_ft'])
df2.head()
```

```python
df2['G_Kpa'] = (df2['G_psi']*101.325)/14.7
df2.head()
```

### 6) Dynamic Young's Modulus (E)

$$ E = 2G ( 1 + PR ) $$

```python
def E(G_psi,PR):
    E= 2*G_psi*(1+PR)
    return E
```

```python
df2['E_psi']=E(df2['G_psi'],df2['PR'])
df2.head()
```

```python
df2['E_Kpa'] = (df2['E_psi']*101.325)/14.7
df2.head()
```

#### Static Young Modulus

Static YM = Dynamic Young Modulus x 0.809

```python
df2['E_static_Kpa'] = 0.809*df2['E_Kpa']
df2.head()
```

### 7) Brittleness index

$$ BRIT_i = 0.0715 {YM}_s - 1.43PR +0.57 $$
$$YMs: Static Young's Modulus (1E6 psi)$$
$$YMd: Dynamic Young's Modulus (1E6 psi)$$
$$BRITi: Brittleness Index (fraction)$$
$$\rho_b: Bulkdensity (g/cc)$$
$$PR: Poisson's Ratio (fraction)$$

```python
def BRIT(E_static_Kpa,PR_static):
    BRIT=0.0715*((E_static_Kpa*14.7)/(101.325*1000000)) - 1.43*PR_static + 0.57
    return BRIT
```

```python
df2['BRIT']= BRIT(df2['E_static_Kpa'],df2['PR_static'])
df2.head()
```

### 8) Vertical Biot's constant (av=0.6+[ma*feffec]) Detournay and Cheng (1993)

$$av = 0.6 + [ma*f_{effec}]$$

```python
def Biot_v(ma,effective_porosity):
    Biot_v=0.6+(ma*effective_porosity)
    return Biot_v
```

```python
df = pd.read_csv("HW6.csv")
```

```python
ma=1
df2['Biot_v']= Biot_v(ma,df['Effective_Porosity'])
df2.head()
```

### 9) Overburden Stress (sz)

```python
def sz(depth, overburden):
    sz = (depth/0.3048) * overburden
    return sz
```

```python
overburden = 1
df2['sz_psi']= sz(df2['DEPTH'],overburden)
df2.head()
```

```python
df2['sz_Kpa']= (df2['sz_psi']*101.321)/14.7
df2.head()
```

### 10) Draw curves for GR logs and calculated parameteres in questions 4,5,6,7

```python
def make_logs_plt(logs2):
    #sort logs by depth
    logs2 = logs2.sort_values(by='DEPTH')
    stop=logs2.DEPTH.min()
    xbot=logs2.DEPTH.max()
    
    fig, ax = plt.subplots(nrows=1,ncols=5, figsize=(20,10))
    ax[0].plot(logs2.GR, logs2.DEPTH, '-', color='darkgreen')
    ax[1].plot(logs2.PR, logs2.DEPTH, '-', color='blue')
    ax[2].plot(logs2.G_Kpa, logs2.DEPTH, '-', color='magenta')
    ax[3].plot(logs2.E_Kpa, logs2.DEPTH, '-', color='black')
    ax[4].plot(logs2.BRIT, logs2.DEPTH, '-', color='red')
    
    for i in range(len(ax)):
        ax[i].set_ylim(stop,xbot)
        ax[i].invert_yaxis()
        ax[i].grid()
        ax[i].locator_params(axis='x', nbins=3)
        
    #set x-axis
    ax[0].set_xlabel("GR")
    ax[0].set_xlim(0,150)
    ax[1].set_xlabel("Dyanmic Poisson's ratio")
    ax[1].set_xlim(logs2.PR.min(),logs2.PR.max())
    ax[2].set_xlabel("Dyanmic Shear Modulus")
    ax[2].set_xlim(logs2.G_Kpa.min(),logs2.G_Kpa.max())    
    ax[3].set_xlabel("Dyanmic Young's Modulus")
    ax[3].set_xlim(logs2.E_Kpa.min(),logs2.E_Kpa.max())    
    ax[4].set_xlabel("Brittleness Index")
    ax[4].set_xlim(0,1)    
    
    #set y-axis
    ax[1].set_yticklabels([])
    ax[2].set_yticklabels([])
    ax[3].set_yticklabels([])
    ax[4].set_yticklabels([])
```

```python
make_logs_plt(df2)
```
